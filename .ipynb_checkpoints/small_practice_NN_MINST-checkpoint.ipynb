```python
import tensorflow as tf 
```

```python
(x_train, y_train), (x_test, y_test) \
    = tf.keras.datasets.mnist.load_data() ## loads in tuple
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
## make dataset to tensor type
x_train = tf.convert_to_tensor(x_train, dtype = tf.float32)/ 255.0
y_train = y_train.reshape(60000,1) ## change the shape for db
y_train = tf.convert_to_tensor(y_train, dtype = tf.float32)
```

```python
## generate dataset for batch learning
db = tf.data.Dataset.from_tensor_slices((x_train, y_train))
```

```python
for step, (x,y) in enumerate(db):
    print(step, x.shape, y, y.shape)
```

```python
print(x_train.shape[1:])
inputs = tf.keras.layers.Input(shape = (28*28))
```

```python
L = tf.keras.layers.Dense(512, activation = "relu")(inputs)
L = tf.keras.layers.Dense(256, activation = "relu")(L)
outputs = tf.keras.layers.Dense(10)(L)
```
